```python
import requests
from bs4 import BeautifulSoup
import smtplib
import time

URL = "https://www.amazon.in/OnePlus-Mirror-Black-128GB-Storage/dp/B085J17VVP/ref=sr_1_1?dchild=1&keywords=oneplus+8t&qid=1618289743&sr=8-1"
HEADERS ={"User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/84.0.4147.89 Safari/537.36"}

#ENTER YOUR REQUIRED PRICE 
Required_price = 30000.00

#ENTER YOUR PERSONAL EMAIL ADDRESS HERE
EMAIL_ADDRESS= "YOUR-EMAIL-ADDRESS@gmail.com"


def compare():
    price = eval(TitlePrice())
    if price>=Required_price:
        diff= price - Required_price
        print("the required price is still {}".format(diff))
    else:
        print("the price of the product is cheaper now ... Hurry up !!!")
        sendmail()


def TitlePrice():
    page= requests.get(URL, headers=HEADERS)
    soup= BeautifulSoup(page.content, "html.parser")
    title = soup.find(id = "productTitle").get_text().strip()
    price= soup.find(id= "priceblock_dealprice").get_text().strip()[1:8]
    price=price[1:]
    price_ar= price.split(",")
    price=''.join(price_ar)
    print()
    print(title)
    print(price)
    return price

#IF YOU ARE USING GMAIL, THEN YOU HAVE TO CHANGE SOME SETTINGS 

#because google does't want to automate the things with other softwares... They want us to send emails from gmail application

#FOLLOW THIS STEPS
#STEP -1 GO TO YOUR GMAIL ACCOUNT AND CLICK ON YOUR PROFILE PICTURE
#STEP -2 click on MANAGE YOUR GOOGLE ACCOUNT
#STEP -3 Go to the SECURITY section (left side of the screen)
#STEP -4 scroll down and click on LESS SECURE APP ACCESS
#STEP -5 TURN ON THE ACCESS



def sendmail():
    subject= "Hey there bud !! the price has been dropped hurry Up !!!!"
    body = "subject:" + subject+'\n\n'+URL
    server = smtplib.SMTP(host='smtp.gmail.com', port = 587)
    server.ehlo()
    server.starttls()
    #TRY TO ENTER YOUR PASSWORD HERE
    server.login(EMAIL_ADDRESS, 'ENTER-PASSWORD-HERE')
    server.sendmail(EMAIL_ADDRESS,EMAIL_ADDRESS, body)
    print("mail sent sucessfully")
    pass 

if __name__=="__main__":
    while True:
        compare()
        time.sleep(40)
```
