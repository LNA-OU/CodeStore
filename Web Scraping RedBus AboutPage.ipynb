```python
# lets do web scraping -- Today we are going to web scrape the redbus about section 
import requests
from bs4 import BeautifulSoup

URL = 'https://www.redbus.in/info/aboutus'
response = requests.get(URL)
Soup = BeautifulSoup(response.content, 'html5lib')
print(Soup.prettify())
```

```python
# now we had managed to get the whole HTML page, the next step is to get what we want ..

RedbusAbout = Soup.findAll(attrs= {'class': 'Red XCN'})[0].text
print(RedbusAbout)
type(RedbusAbout)
```

```python
AboutRedBus = Soup.findAll(attrs= {'class': 'western'})[0].text
print(AboutRedBus)
```
